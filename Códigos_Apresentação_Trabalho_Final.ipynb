```python
# Script 3 – Filtrar Lista de Empresas ISE B3 2023

# Instalando bibliotecas
!pip install beautifulsoup4
!pip install requests

#Importando Bibliotecas
import re
import requests
from bs4 import BeautifulSoup

# Definindo URL da página
url = "https://iseb3.com.br/carteiras-e-questionarios"

# Fazendo a requisição para a página
response = requests.get(url)

if response.status_code == 200:
    # Extraindo o conteúdo HTML da página
    html_content = response.content

    soup = BeautifulSoup(html_content, "html.parser")

    # Encontrando o parágrafo com os nomes das empresas
    paragraph = soup.find("p", string=lambda text: text and text.startswith("Aeris"))

    # Extraindo o texto do parágrafo
    paragraph_text = paragraph.get_text(strip=True)

    # Obtendo o trecho de texto contendo as empresas
    companies_text = re.search(r"Aeris(.*?Weg\.?)", paragraph_text).group(0)

    # Separando as empresas em uma lista
    companies_list = re.split(r",\s*(?=.{2,}$)", companies_text)

    # Tratando uma condição de exceção da última empresa
    companies_list[-1] = companies_list[-1].rstrip(".,")
    last_company = companies_list[-1].split(" e ")
    companies_list = companies_list[:-1] + last_company

    # Exibir os nomes das empresas por linha
    for company in companies_list:
        print(company.rstrip(".,"))
else:
    print("Falha ao obter a página:", response.status_code)
```

```python
# Script 6 – Plotar Gráfico com a Variação das Ações - Plotly

# Instalando bibliotecas
!pip install pandas
!pip install yfinance
!pip install plotly

#Importando Bibliotecas
import yfinance as yf
import datetime
import plotly.graph_objects as go

# Definindo as empresas e suas ações
empresas = {
    'WEG': 'WEGE3.SA',
    'Suzano': 'SUZB3.SA',
    'Klabin': 'KLBN11.SA',
    'Raízen': 'RAIZ4.SA',
    'Natura': 'NTCO3.SA'
}

# Definindo o período de 5 anos
data_inicio = datetime.datetime.now() - datetime.timedelta(days=5*365)
data_fim = datetime.datetime.now()

fig = go.Figure()

for empresa, ticker in empresas.items():
    # Obtendo os dados das ações
    acao = yf.Ticker(ticker)
    dados = acao.history(start=data_inicio, end=data_fim)

    # Calculando a variação percentual
    preco_inicial = dados['Close'].iloc[0]
    preco_final = dados['Close'].iloc[-1]
    variacao_percentual = (preco_final - preco_inicial) / preco_inicial * 100

    # Adicionando uma linha para cada empresa
    fig.add_trace(go.Scatter(
        x=dados.index,
        y=dados['Close'],
        mode='lines',
        name=f'{empresa} (Variação: {variacao_percentual:.2f}%)'  # adicionando variação percentual no nome
    ))

# Configurando o layout do gráfico
fig.update_layout(
    title='Variação das Ações',
    xaxis_title='Data',
    yaxis_title='Preço',
    legend=dict(
        orientation='h',
        yanchor='bottom',
        y=-0.3,
        xanchor='right',
        x=1
    )
)

# Exibir o gráfico
fig.show()
```

```python
# Script 7 – Plotar Gráfico do Histórico de Preço/Lucro das Ações

# Instalando bibliotecas
!pip install yfinance
!pip install matplotlib

#Importando Bibliotecas
import yfinance as yf
import matplotlib.pyplot as plt

#Definindo as ações
tickers = ['WEGE3.SA', 'SUZB3.SA', 'KLBN11.SA', 'RAIZ4.SA', 'NTCO3.SA']
pl_data = {}  # Criando um espaço para rmazenar os dados de P/L

#Coletando os Dados
for ticker in tickers:
    acao = yf.Ticker(ticker)
    dados = acao.history(period='5y')
    ultimo_precos = dados['Close']
    lucro_por_acao = acao.info['trailingEps']
    pl_ratios = ultimo_precos / lucro_por_acao

    pl_data[ticker] = pl_ratios

# Plotar o gráfico de P/L
plt.figure(figsize=(10, 6))
for ticker, pl in pl_data.items():
    plt.plot(pl.index, pl.values, label=ticker)

plt.title("Histórico de Preço/Lucro")
plt.xlabel("Data")
plt.ylabel("P/L")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Script 7.1 – Plotar Gráfico do Preço/Lucro atual

# Instalando bibliotecas
!pip install yfinance
!pip install matplotlib

#Importando Bibliotecas
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['WEGE3.SA', 'SUZB3.SA', 'KLBN11.SA', 'RAIZ4.SA', 'NTCO3.SA']
pl_ratios = []

for ticker in tickers:
    # Obtendo os dados da ação
    acao = yf.Ticker(ticker)
    dados = acao.history(period='1d')

    # Calculando o preço sobre lucro (P/L)
    ultimo_preco = dados['Close'][-1]
    lucro_por_acao = acao.info['trailingEps']
    pl_ratio = ultimo_preco / lucro_por_acao
    pl_ratios.append(pl_ratio)

    # Exibindo o resultado
    print(f'O P/L da ação {ticker} é: {pl_ratio:.2f}')

# Criando um gráfico de barras com os valores de P/L
fig, ax = plt.subplots()
ax.bar(tickers, pl_ratios)

# Personalizando o gráfico
ax.set_xlabel('Empresa')
ax.set_ylabel('P/L')
ax.set_title('Relação Preço/Lucro das Ações')

# Exibir o gráfico
plt.show()
```

```python
# Script 9 – Plotar Gráfico do Retorno Acumulado das Ações

# Instalando bibliotecas
!pip install pandas
!pip install yfinance
!pip install matplotlib

#Importando Bibliotecas
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Definindo lista de códigos das ações das empresas
empresas_brasileiras = ['WEGE3.SA', 'SUZB3.SA', 'KLBN11.SA', 'RAIZ4.SA', 'NTCO3.SA']

# Definindo período de análise
data_inicio = '2020-01-01'
data_fim = '2022-12-31'

# Baixando dados das ações
dados_brasileiros = yf.download(empresas_brasileiras, start=data_inicio, end=data_fim)['Adj Close']

# Calculando retornos diários
retornos_brasileiros = dados_brasileiros.pct_change()

# Calculando retornos acumulados
retorno_acumulado_brasileiro = (1 + retornos_brasileiros).cumprod()

# Plotar gráfico de retornos acumulados
retorno_acumulado_brasileiro.plot(title='Retorno Acumulado das Ações')

# Exibir gráfico
plt.show()
```

```python
# Script 9.1 – Plotar Gráfico do Retorno Acumulado das Ações - Plotly

# Instalando bibliotecas
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go

# Definindo lista de códigos das ações das empresas
empresas_brasileiras = ['WEGE3.SA', 'SUZB3.SA', 'KLBN11.SA', 'RAIZ4.SA', 'NTCO3.SA']

# Definindo período de análise
data_inicio = '2020-01-01'
data_fim = '2022-12-31'

# Baixando dados das ações
dados_brasileiros = yf.download(empresas_brasileiras, start=data_inicio, end=data_fim)['Adj Close']

# Calculando retornos diários
retornos_brasileiros = dados_brasileiros.pct_change()

# Calculando retornos acumulados
retorno_acumulado_brasileiro = (1 + retornos_brasileiros).cumprod()

# Criando um gráfico de linha para cada ação
fig = go.Figure()
for empresa in retorno_acumulado_brasileiro.columns:
    fig.add_trace(go.Scatter(
        x=retorno_acumulado_brasileiro.index,
        y=retorno_acumulado_brasileiro[empresa],
        mode='lines',
        name=empresa
    ))

# Configurando o layout do gráfico
fig.update_layout(
    title='Retorno Acumulado das Ações',
    xaxis_title='Período',
    yaxis_title='Retorno Acumulado',
)

# Exibir o gráfico
fig.show()
```

```python
# Script 9.1 – Plotar Gráfico de Correlação das Ações

# Instalando Bibliotecas
!pip install pandas_datareader
!pip install seaborn

# Importando Bibliotecas
import datetime as dt
import seaborn as sns
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
import yfinance as yf

# Baixando dados das Ações para um DataFrame
yf.pdr_override()
start_date = dt.datetime(2020, 1, 1)
end_date = dt.datetime.now()
tickers = ["WEGE3.SA", "SUZB3.SA", "KLBN11.SA", "RAIZ4.SA", "NTCO3.SA"]
data = pdr.get_data_yahoo(tickers, start_date, end_date)["Adj Close"]

# Calculando Retornos Diários
returns_data = data.pct_change().dropna()

# Criando uma Matriz de Correlação
correlation = returns_data.corr()

# Plotar o Mapa de Calor de Correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Heatmap de Correlação das Ações")
plt.show()
     
```
