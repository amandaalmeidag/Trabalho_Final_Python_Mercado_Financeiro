```python
import re
import requests
from bs4 import BeautifulSoup

# URL da página
url = "https://iseb3.com.br/carteiras-e-questionarios"

# Faz a requisição GET para a página
response = requests.get(url)

# Verifica se a requisição foi bem-sucedida
if response.status_code == 200:
    # Extrai o conteúdo HTML da resposta
    html_content = response.content

    # Cria um objeto BeautifulSoup para fazer o parsing do HTML
    soup = BeautifulSoup(html_content, "html.parser")

    # Encontra o parágrafo que contém os nomes das empresas
    paragraph = soup.find("p", string=lambda text: text and text.startswith("Aeris"))

    # Extrai o texto do parágrafo
    paragraph_text = paragraph.get_text(strip=True)

    # Obtém o trecho de texto contendo as empresas
    companies_text = re.search(r"Aeris(.*?Weg\.?)", paragraph_text).group(0)

    # Separa as empresas em uma lista
    companies_list = re.split(r",\s*(?=.{2,}$)", companies_text)

    # Trata a exceção da última empresa
    companies_list[-1] = companies_list[-1].rstrip(".,")
    last_company = companies_list[-1].split(" e ")
    companies_list = companies_list[:-1] + last_company

    # Exibe os nomes das empresas por linha
    for company in companies_list:
        print(company.rstrip(".,"))
else:
    print("Falha ao obter a página:", response.status_code)
```

```python
#Variação das Ações nos últimos anos - Plotly
!pip install plotly
import yfinance as yf
import datetime
import plotly.graph_objects as go

# Definir as empresas e seus respectivos tickers
empresas = {
    'WEG': 'WEGE3.SA',
    'Suzano': 'SUZB3.SA',
    'Klabin': 'KLBN11.SA',
    'Raízen': 'RAIZ4.SA',
    'Natura': 'NTCO3.SA'
}

# Definir o período de 5 anos
data_inicio = datetime.datetime.now() - datetime.timedelta(days=5*365)
data_fim = datetime.datetime.now()

fig = go.Figure()

for empresa, ticker in empresas.items():
    # Obter os dados da ação utilizando o yfinance
    acao = yf.Ticker(ticker)
    dados = acao.history(start=data_inicio, end=data_fim)

    # Calcular a variação percentual
    preco_inicial = dados['Close'].iloc[0]
    preco_final = dados['Close'].iloc[-1]
    variacao_percentual = (preco_final - preco_inicial) / preco_inicial * 100

    # Adicionar um trace ao gráfico
    fig.add_trace(go.Scatter(
        x=dados.index,
        y=dados['Close'],
        mode='lines',
        name=empresa
    ))

# Configurar o layout do gráfico
fig.update_layout(
    title='Variação das Ações nos Últimos 5 Anos',
    xaxis_title='Data',
    yaxis_title='Preço de Fechamento',
    legend=dict(
        orientation='h',
        yanchor='bottom',
        y=-0.3,
        xanchor='right',
        x=1
    )
)

# Exibir o gráfico
fig.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['WEGE3.SA', 'SUZB3.SA', 'KLBN11.SA', 'RAIZ4.SA', 'NTCO3.SA']
pl_data = {}  # Dicionário para armazenar os dados de P/L

for ticker in tickers:
    acao = yf.Ticker(ticker)
    dados = acao.history(period='5y')
    ultimo_precos = dados['Close']
    lucro_por_acao = acao.info['trailingEps']
    pl_ratios = ultimo_precos / lucro_por_acao

    pl_data[ticker] = pl_ratios

# Plotar o gráfico
plt.figure(figsize=(10, 6))
for ticker, pl in pl_data.items():
    plt.plot(pl.index, pl.values, label=ticker)

plt.title("Histórico do P/L das Empresas nos Últimos 5 Anos")
plt.xlabel("Data")
plt.ylabel("P/L")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Preço/Lucro atual
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['WEGE3.SA', 'SUZB3.SA', 'KLBN11.SA', 'RAIZ4.SA', 'NTCO3.SA']
pl_ratios = []

for ticker in tickers:
    # Obter os dados da ação utilizando o yfinance
    acao = yf.Ticker(ticker)
    dados = acao.history(period='1d')

    # Calcular o preço sobre lucro (P/L)
    ultimo_preco = dados['Close'][-1]
    lucro_por_acao = acao.info['trailingEps']
    pl_ratio = ultimo_preco / lucro_por_acao
    pl_ratios.append(pl_ratio)

    # Exibir o resultado
    print(f'O P/L da ação {ticker} é: {pl_ratio:.2f}')


# Criar um gráfico de barras com os valores de P/L
fig, ax = plt.subplots()
ax.bar(tickers, pl_ratios)

# Personalizar o gráfico
ax.set_xlabel('Empresa')
ax.set_ylabel('P/L')
ax.set_title('Relação Preço/Lucro das Ações')

# Exibir o gráfico
plt.show()
```

```python
#Retorno Acumulado
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Definir lista de símbolos das empresas brasileiras
empresas_brasileiras = ['WEGE3.SA', 'SUZB3.SA', 'KLBN11.SA', 'RAIZ4.SA', 'NTCO3.SA']

# Definir período de análise
data_inicio = '2020-01-01'
data_fim = '2022-12-31'

# Baixar dados das ações usando yfinance
dados_brasileiros = yf.download(empresas_brasileiras, start=data_inicio, end=data_fim)['Adj Close']

# Calcular retornos diários
retornos_brasileiros = dados_brasileiros.pct_change()

# Calcular retornos acumulados
retorno_acumulado_brasileiro = (1 + retornos_brasileiros).cumprod()

# Plotar gráfico de retornos acumulados
retorno_acumulado_brasileiro.plot(title='Retorno Acumulado - Empresas Brasileiras')

# Exibir gráfico
plt.show()
```

```python
#Retorno do Investimento
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Definir a lista de símbolos das empresas
empresas = ['WEGE3.SA', 'SUZB3.SA', 'KLBN11.SA', 'RAIZ4.SA', 'NTCO3.SA']

# Criar um dicionário para armazenar os dados financeiros de cada empresa
dados_empresas = {}

# Obter os dados financeiros para cada empresa
for empresa in empresas:
    # Usar a biblioteca yfinance para obter os dados históricos da ação
    dados = yf.download(empresa, start='2013-01-01', end='2023-06-11')
    # Armazenar os dados no dicionário
    dados_empresas[empresa] = dados

# Criar um dataframe para armazenar o retorno de investimento
retorno_investimento = pd.DataFrame(columns=['Empresa', 'Retorno'])

# Calcular o retorno de investimento para cada empresa
for empresa, dados in dados_empresas.items():
    # Verificar se há dados disponíveis para a empresa
    if len(dados) > 0:
        # Calcular o retorno de investimento como a diferença percentual entre o preço de fechamento atual e o preço de fechamento inicial
        retorno = (dados['Close'][-1] / dados['Close'][0]) - 1
        # Adicionar o retorno de investimento ao dataframe
        retorno_investimento = retorno_investimento.append({'Empresa': empresa, 'Retorno': retorno}, ignore_index=True)
        
        # Plotar o gráfico de preço de fechamento da ação
        plt.plot(dados.index, dados['Close'], label=empresa)
    else:
        print(f"Dados não disponíveis para a empresa: {empresa}")

# Configurar a legenda e os rótulos do gráfico
plt.legend()
plt.title('Preço de Fechamento das Ações')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')

# Exibir o gráfico interativo
plt.show()

# Exibir o resultado do retorno de investimento
print(retorno_investimento)
```

```python
# PASSO 2 - Importando Bibliotecas
import datetime as dt
import seaborn as sns
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
import yfinance as yf

# PASSO 3 - Baixando dados das Ações para um DataFrame
yf.pdr_override()
start_date = dt.datetime(2020, 1, 1)
end_date = dt.datetime.now()
tickers = ["WEGE3.SA", "SUZB3.SA", "KLBN11.SA", "RAIZ4.SA", "NTCO3.SA"]
data = pdr.get_data_yahoo(tickers, start_date, end_date)["Adj Close"]

# PASSO 4 - Calcular Retornos Diários
returns_data = data.pct_change().dropna()

# PASSO 5 - Criar uma Matriz de Correlação
correlation = returns_data.corr()

# PASSO 6 - Plotar o Heatmap de Correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.show()
```
