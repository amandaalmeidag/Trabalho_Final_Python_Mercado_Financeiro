```python
from google.colab import files
import pandas as pd
import io
import matplotlib.pyplot as plt

# Carrega o arquivo
uploaded = files.upload()

# Agora vamos ler o arquivo carregado usando pandas
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(name=fn, length=len(uploaded[fn])))
  df = pd.read_csv(io.StringIO(uploaded[fn].decode('utf-8')), sep=';')

# Exibe as primeiras linhas do DataFrame
print(df.head())

# Cria um novo DataFrame onde cada linha é um ano e cada coluna é a variação do ISE para esse ano
df_melt = df.melt(id_vars='Ano', var_name='Year_Compared', value_name='ISE_Change')

# Converte as colunas de ano para inteiros
df_melt['Ano'] = df_melt['Ano'].astype(int)
df_melt['Year_Compared'] = df_melt['Year_Compared'].astype(int)

# Remove as linhas onde Year_Compared >= Ano, já que essas células estavam vazias na tabela original
df_melt = df_melt[df_melt['Ano'] > df_melt['Year_Compared']]

# Plota o gráfico de linha da variação do ISE por ano
plt.figure(figsize=(12, 6))
for year in df_melt['Ano'].unique():
  df_year = df_melt[df_melt['Ano'] == year]
  plt.plot(df_year['Year_Compared'], df_year['ISE_Change'], marker='o', label=year)

plt.title('Variação do Índice ISE por ano')
plt.xlabel('Ano')
plt.ylabel('Variação do ISE')
plt.legend(title='Ano de referência')

plt.show()
```

```python
from google.colab import files
import pandas as pd
import io
import matplotlib.pyplot as plt

# Carrega o arquivo
uploaded = files.upload()

# Agora vamos ler o arquivo carregado usando pandas
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(name=fn, length=len(uploaded[fn])))
  df = pd.read_csv(io.StringIO(uploaded[fn].decode('utf-8')), sep=';')

# Exibe as primeiras linhas do DataFrame
print(df.head())

# Supondo que a coluna de datas é chamada 'Date' e está no formato 'YYYY-MM-DD'
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Calcula a variação média do ISE por ano (altere 'ISE' se o nome da sua coluna for diferente)
df_grouped = df.groupby('Year')['ISE'].mean().pct_change()

# Plota o gráfico de linha da variação média do ISE por ano
df_grouped.plot(kind='line', figsize=(10, 5))

plt.title('Taxa média de variação do Índice ISE por ano')
plt.xlabel('Ano')
plt.ylabel('Taxa média de variação')

plt.show()
```
